Dados os valores de um depósito fixo mensal e um montante desejado, crie uma classe para determinar quantos meses serão necessários para acumular o montante desejado, considerando juros mensais de 0,5%.

```java
import java.util.Scanner;

Scanner LER = new Scanner(System.in);

double depositoMensal;
double montanteDesejado;
int meses = 1;
double c = 0;
          
System.out.print("Informe o valor do deposito fixo mensal: ");
depositoMensal = LER.nextDouble();
          
System.out.print("\nInforme o montante desejado: ");
montanteDesejado = LER.nextDouble();

c = depositoMensal;
          
while (c < montanteDesejado){
              
    c = c + (depositoMensal * 1.05);
    meses = meses + 1;
          
}
          
System.out.printf("%nSerá necessario %s meses até chegar no monte desejado.%n", meses);
        
System.out.printf("O Valor adquirido nesses %d meses foi de: R$%.2f", meses, c);
```
